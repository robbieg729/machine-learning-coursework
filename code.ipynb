```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier


##Data frame manipulation
df = pd.read_csv("data.csv") # Loading data
print("Number of individuals =", np.shape(df)[0]) # 1071 rows
df.info() # Each column has 1071 non-null values, so no null values in entire data frame

print("Age range =", np.max(df["age"]) - np.min(df["age"])) # see unusual range of 44835 for ages, so some data must be bad input
print("Max age =", np.max(df["age"]))
print("Min age =", np.min(df["age"]))
plt.hist(df.values[:, 36], bins=np.arange(0, 200, 10)) # ages histogram
plt.title("Histogram of ages")
plt.ylabel("Frequency", fontsize=15)
plt.xlabel("Ages", fontsize=15)
plt.show()
plt.cla()
plt.clf()

print("Gender counts =\n", df["gender"].value_counts(), "\n") # See 5 values of 0, all the rest 1, 2, 3

plt.hist(df.values[:, 38], bins=np.arange(0, 105, 5)) # Accuracies histogram
plt.title("Histogram of accuracies")
plt.ylabel("Frequency", fontsize=15)
plt.xlabel("Accuracies", fontsize=15)
plt.show()
plt.cla()
plt.clf()

# Delete all rows with age > 100, gender != 1, 2, or 3, and accuracy < 75
df = pd.DataFrame(np.delete(df.values, np.where((df.values[:, 36] > 100) | (df.values[:, 37] < 1) | (df.values[:, 38] < 75)), axis=0), columns=df.columns) # Recreate data frame

self_defeating_question_indices = np.arange(start=3, stop=32, step=4) # Indices of answers used to calculate self-defeating scale score
for column in df.columns[self_defeating_question_indices]: 
    print(df[column].value_counts(), "\n") # Check counts for answers to questions, see several missing values (-1)

questions = df.values[:, self_defeating_question_indices] # Extract answers to self-defeating questions into numpy array
imputer = SimpleImputer(missing_values=-1, strategy="most_frequent") 
questions = imputer.fit_transform(questions) # replace missing answers with most frequent answer to that question

answer_values = np.unique(questions) # List of unique values in questions
print("New answer values =", answer_values) # Now only values of 1, 2, 3, 4, 5

new_scale_scores = np.around(np.sum(questions, axis=1) / 8, decimals=1) # Calculate new self-defeating scale scores

y = np.where(new_scale_scores <= 2, 0,
                np.where(new_scale_scores <= 3, 1, 
                np.where(new_scale_scores <= 4, 2, 
                np.where(new_scale_scores <= 5, 3, 3)))) # Put scale scores into classes

print("Target classes list =", y) # See a list of values between 0 and 3 inclusive

X = np.hstack((questions[:, [0, 2, 4, 6]], df.values[:, 36:38])) # Answers to 4 questions, with ages and genders

print("New number of individuals =", np.shape(X)[0]) # Reduced from 1071 instances to 969 instances

print("Data set =", X)

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
print("Data set fully scaled =", X) # See data is fully standardized

def tune_hyperparameters(svc, hyperparams, X, y):
    clf = GridSearchCV(svc, hyperparams, cv=10, scoring="accuracy")
    clf.fit(X, y)
    return clf.best_params_

def get_cross_validate_score(clf, X, y):
    cv = cross_validate(clf, X, y, scoring="accuracy", cv=10)
    return np.mean(cv["test_score"])

def get_predictions(clf, X_train, y_train, X_test, y_test, clf_type):
    clf.fit(X_train, y_train)
    pred = clf.predict(X_test)
    pred_probas = clf.predict_proba(X_test)
    print(clf_type + " test set accuracy =", accuracy_score(y_test, pred))
    plot_confusion_matrix(clf, X_test, y_test)
    plt.title(clf_type + " confusion matrix")
    plt.show()
    return pred, pred_probas

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=9) # Constant 90/10 train-test split
print("Individual 4 from KNN and Logistic Regression test set =", np.append(X_test[4], y_test[4]))

y_uniq, counts = np.unique(y_train, return_counts=True) # Class counts in training set
plt.bar([0, 0.5, 1, 1.5], counts, width=0.5, color=["r", "b", "g", "y"], tick_label=["1 \u2264 s \u2264 2 (0)", "2 < s \u2264 3 (1)", "3 < s \u2264 4 (2)", "4 < s \u2264 5 (3)"])
plt.xlabel("Scale score s (class label)", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Self-defeating scale score counts with respect to class for training set")
plt.show()

### Logistic Regression
log_reg_hyperparams = {"penalty": ["l1", "l2", "elasticnet"], "solver": ["liblinear", "saga", "lbfgs", "newton-cg", "sag"]}
log_reg_best_params = tune_hyperparameters(LogisticRegression(max_iter=10000), log_reg_hyperparams, X, y) # Get a few warnings as some combinations are not possible
print("Logistic Regression optimal hyperparameters =", log_reg_best_params)
log_reg = LogisticRegression(penalty=log_reg_best_params["penalty"], solver=log_reg_best_params["solver"])
log_reg_accuracy = get_cross_validate_score(log_reg, X, y)
print("Logistic Regression CV accuracy =", log_reg_accuracy)
log_reg_pred, log_reg_pred_probas = get_predictions(log_reg, X_train, y_train, X_test, y_test, "Logistic Regression")
print("Logistic Regression probabilities =", log_reg_pred_probas)

### K-Neighbors
knn_hyperparams = {"n_neighbors": np.arange(1, 20), "metric": ["euclidean", "minkowski", "chebyshev", "manhattan"], "algorithm": ["ball_tree", "kd_tree", "brute", "auto"]}
knn_best_params = tune_hyperparameters(KNeighborsClassifier(), knn_hyperparams, X, y)
print("KNN optimal hyperparameters =", knn_best_params)
knn = KNeighborsClassifier(n_neighbors=knn_best_params["n_neighbors"], metric=knn_best_params["metric"], algorithm=knn_best_params["algorithm"])
knn_accuracy = get_cross_validate_score(knn, X, y)
print("K-Nearest Neighbors CV accuracy =", knn_accuracy)
knn_pred, knn_pred_probas = get_predictions(knn, X_train, y_train, X_test, y_test, "K-Nearest Neighbors")
print("K-Nearest Neighbors probabilities =", knn_pred_probas)

X = scaler.inverse_transform(X)
gpc_scaler = StandardScaler()
gpc_scaler.fit(np.reshape(X[:, 4], (-1, 1)))
X[:, 4] = np.reshape(gpc_scaler.transform(np.reshape(X[:, 4], (-1, 1))), (1, -1))
print("Data set with only ages scaled =", X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=9) # Get exact same 90/10 train-test split
print("Individual 4 from GPC test set =", np.append(X_test[4], y_test[4]))

### Gaussian Process
gaussian = GaussianProcessClassifier()
gaussian_accuracy = get_cross_validate_score(gaussian, X, y)
print("Gaussian CV accuracy =", gaussian_accuracy)
gaussian_pred, gaussian_pred_probas = get_predictions(gaussian, X_train, y_train, X_test, y_test, "Gaussian Process")
print("Gaussian probabilities =", gaussian_pred_probas)
```
